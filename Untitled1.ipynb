```python
import pandas as pd
from rdkit import Chem
from rdkit import RDLogger
from tqdm import tqdm

# Disable RDKit warnings (optional)
RDLogger.DisableLog('rdApp.*')

# === Canonicalize and remove explicit H safely ===
def canonicalize_smiles_remove_H(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None
        mol = Chem.RemoveHs(mol)            # Remove explicit hydrogens
        Chem.SanitizeMol(mol)               # Sanitize (check valence, aromatics, etc.)
        return Chem.MolToSmiles(mol, canonical=True)  # Return canonical SMILES
    except Exception:
        return None  # Return None for any processing error

# === Process CSV file with SMILES ===
def process_smiles_file(input_csv, output_csv, smiles_col='Compounds', label_col='BBB-Class'):
    # Load CSV
    df = pd.read_csv(input_csv)
    
    # Validate column names
    if smiles_col not in df.columns or label_col not in df.columns:
        raise ValueError(f"Input CSV must contain columns: '{smiles_col}' and '{label_col}'")
    
    # Show progress bar
    tqdm.pandas()
    
    # Canonicalize and clean SMILES
    df['canonical_smiles'] = df[smiles_col].progress_apply(canonicalize_smiles_remove_H)
    
    # Filter valid SMILES
    df['valid'] = df['canonical_smiles'].notnull()
    valid_df = df[df['valid']].drop(columns=['valid'])
    
    # Save to output
    valid_df.to_csv(output_csv, index=False)
    
    print(f"✅ Processed {len(df)} SMILES. Valid SMILES saved: {len(valid_df)} to '{output_csv}'")
    return valid_df

# === Example Usage ===
if __name__ == '__main__':
    input_file = 'BBBData.csv'                         # Replace with your file
    output_file = 'BBBData_canonical_no_explicitH.csv' # Output file
    processed_data = process_smiles_file(input_file, output_file, smiles_col='Compounds', label_col='BBB-Class')
```

```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.ML.Descriptors import MoleculeDescriptors
from rdkit.Chem import Descriptors, MACCSkeys
from rdkit.DataStructs.cDataStructs import ConvertToNumpyArray

# Load your CSV file
df = pd.read_csv('BBBData_canonical_no_explicitH.csv')

# Check required columns
if 'canonical_smiles' not in df.columns or 'BBB-Class' not in df.columns:
    raise ValueError("CSV must contain 'canonical_smiles' and 'BBB-Class' columns")

# === RDKit Molecular Descriptors ===
descriptor_names = [desc[0] for desc in Descriptors._descList]
calculator = MoleculeDescriptors.MolecularDescriptorCalculator(descriptor_names)

# Function to calculate descriptors
def calc_descriptors(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return [np.nan] * len(descriptor_names)
    else:
        return calculator.CalcDescriptors(mol)

# === MACCS Fingerprints ===
MACCS_FP_SIZE = 167  # MACCS has 167 bits

def compute_maccs_fingerprint(mol):
    fp = MACCSkeys.GenMACCSKeys(mol)
    arr = np.zeros((MACCS_FP_SIZE,), dtype=int)
    ConvertToNumpyArray(fp, arr)
    return arr.tolist()

# Process molecules and calculate features
desc_list = []
maccs_list = []
valid_smiles = []
labels = []

for idx, row in df.iterrows():
    smiles = row['canonical_smiles']
    bbb_class = row['BBB-Class']
    mol = Chem.MolFromSmiles(smiles)
    
    if mol is None:
        # Skip invalid SMILES
        continue

    # Calculate descriptors and fingerprints
    descriptors = calc_descriptors(smiles)
    maccs_fp = compute_maccs_fingerprint(mol)
    
    desc_list.append(descriptors)
    maccs_list.append(maccs_fp)
    valid_smiles.append(smiles)
    labels.append(bbb_class)

# Convert to DataFrames
desc_df = pd.DataFrame(desc_list, columns=descriptor_names)
maccs_columns = [f"MACCS_{i}" for i in range(MACCS_FP_SIZE)]
maccs_df = pd.DataFrame(maccs_list, columns=maccs_columns)

# Combine all into a final DataFrame
final_df = pd.DataFrame({
    'canonical_smiles': valid_smiles,
    'BBB-Class': labels
})
final_df = pd.concat([final_df, desc_df, maccs_df], axis=1)

# === Show and save rows with NaN values ===
nan_rows = final_df[final_df.isna().any(axis=1)]
print("Rows with NaN values:")
print(nan_rows)
nan_rows.to_csv('rows_with_nans.csv', index=False)

# === Drop NaNs and save cleaned data ===
final_df.dropna(inplace=True)
final_df.to_csv('molecular_descriptors_with_maccs.csv', index=False)

# Preview result
print("\nFinal cleaned dataset (first 5 rows):")
print(final_df.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from imblearn.over_sampling import SMOTE
from collections import Counter
import joblib

# Step 1: Separate features and target
X = final_df.drop(columns=["BBB-Class", "canonical_smiles"])
y = final_df["BBB-Class"]

# Step 2: Handle Missing Values
imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Step 3: Replace infinite values with NaN and re-impute
X_imputed.replace([np.inf, -np.inf], np.nan, inplace=True)
X_imputed = pd.DataFrame(imputer.fit_transform(X_imputed), columns=X.columns)

# Optional: Clip extreme values
X_imputed = X_imputed.clip(-1e6, 1e6)

# Step 4: Noise Removal (Outlier Detection)
iso = IsolationForest(contamination=0.2, random_state=42)
outliers = iso.fit_predict(X_imputed)
mask = outliers != -1  # keep only non-outliers

X_clean = X_imputed[mask].reset_index(drop=True)
y_clean = y[mask].reset_index(drop=True)
joblib.dump(imputer , 'imputer.pkl')

# Step 5: Feature Scaling
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_clean), columns=X_clean.columns)
joblib.dump(scaler  , 'scaler.pkl')

# No feature selection - use all features
X_selected = X_scaled.copy()


# Check for NaNs before SMOTE
assert not X_selected.isnull().any().any(), "NaNs found before SMOTE!"

# Step 6: Data Balancing using SMOTE
print(f"Class distribution before SMOTE: {Counter(y_clean)}")
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X_selected, y_clean)
print(f"Class distribution after SMOTE: {Counter(y_balanced)}")

# Combine back with target
preprocessed_data = pd.concat([pd.DataFrame(X_balanced, columns=X_selected.columns), 
                               pd.Series(y_balanced, name="BBB-Class")], axis=1)

# Optional: Save to CSV
preprocessed_data.to_csv("molecular_descriptors_with_maccs.csv", index=False)

print("Preprocessing and balancing complete.")
print(f"Final feature shape: {X_balanced.shape}")
print(f"Final dataset shape: {preprocessed_data.shape}")
```

```python
from sklearn.svm import SVC
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, roc_curve
)
import seaborn as sns
import matplotlib.pyplot as plt
import joblib

# List of SVM kernel types to evaluate
kernels = [
    ("linear", SVC(kernel='linear', probability=True)),
    ("sigmoid", SVC(kernel='sigmoid', probability=True)),
    ("rbf", SVC(kernel='rbf', probability=True)),
    ("poly (degree=3)", SVC(kernel='poly', degree=3, probability=True))
]

for name, model in kernels:
    print(f"\n===== Kernel: {name} =====")

    # Fit and predict
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    # Compute metrics
    acc = round(accuracy_score(y_test, y_pred) * 100, 2)
    prec = round(precision_score(y_test, y_pred) * 100, 2)
    rec = round(recall_score(y_test, y_pred) * 100, 2)
    f1 = round(f1_score(y_test, y_pred) * 100, 2)
    auc = round(roc_auc_score(y_test, y_prob) * 100, 2)

    # === Create 1 figure with 3 sections ===
    fig, axs = plt.subplots(1, 3, figsize=(18, 5))  # 1 row, 3 columns

    # 1. Metrics summary (leftmost)
    metric_text = (
        f"Accuracy:  {acc} %\n"
        f"Precision: {prec} %\n"
        f"Recall:    {rec} %\n"
        f"F1 Score:  {f1} %\n"
        f"ROC-AUC:   {auc} %"
    )
    axs[0].axis('off')
    axs[0].text(0.05, 0.5, metric_text, fontsize=14, verticalalignment='center')

    # 2. Confusion matrix (middle)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges', ax=axs[1])
    axs[1].set_title(f'Confusion Matrix\n({name})')
    axs[1].set_xlabel('Predicted')
    axs[1].set_ylabel('Actual')

    # 3. ROC curve (rightmost)
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    axs[2].plot(fpr, tpr, label=f'AUC = {auc}%')
    axs[2].plot([0, 1], [0, 1], 'k--')
    axs[2].set_title(f'ROC Curve\n({name})')
    axs[2].set_xlabel("False Positive Rate")
    axs[2].set_ylabel("True Positive Rate")
    axs[2].legend(loc='lower right')
    axs[2].grid(True)

    # Final adjustments
    fig.suptitle(f'SVM Results - {name}', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

    # Save image
    image_filename = f"svm_{name.replace(' ', '_').replace('(', '').replace(')', '').replace('=', '')}.png"
    fig.savefig(image_filename, dpi=300)
    print(f" Saved image: {image_filename}")

    # Save model + metrics
    model_with_metrics = {
        "model": model,
        "metrics": {
            "Accuracy (%)": acc,
            "Precision (%)": prec,
            "Recall (%)": rec,
            "F1 Score (%)": f1,
            "ROC_AUC (%)": auc,
            "Kernel": name
        }
    }
    pkl_filename = f"supvm_{name.replace(' ', '_').replace('(', '').replace(')', '').replace('=', '')}.pkl"
    joblib.dump(model_with_metrics, pkl_filename)
    print(f" Model saved to: {pkl_filename}")
```

```python
import joblib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve, roc_auc_score

# === Load model metrics ===
model_files = {
    "Logistic Regression": "logistic_with_metrics.pkl",
    "Tunned SVM": "svm_with_metrics.pkl",
    "Tunned KNN": "knn_with_metrics.pkl",
    "Random Forest": "rf_grid_with_metrics.pkl",
    "XGBoost": "xgb_with_metrics.pkl",
    "LightGBM": "lgbm_with_metrics.pkl",
    "Stacking Ensemble": "stacking_with_metrics.pkl"
}

# === Extract metrics ===
all_data = []
for model_name, file in model_files.items():
    try:
        bundle = joblib.load(file)
        metrics = bundle["metrics"]
        all_data.append({
            "Model": model_name,
            "Accuracy": metrics["Accuracy (%)"],
            "Precision": metrics["Precision (%)"],
            "Recall": metrics["Recall (%)"],
            "F1 Score": metrics["F1 Score (%)"],
            "ROC-AUC": metrics["ROC_AUC (%)"]
        })
    except Exception as e:
        print(f"Error loading {file}: {e}")

# === Create DataFrames ===
df = pd.DataFrame(all_data).set_index("Model")
df_sorted = df.sort_values(by="Accuracy", ascending=False)

# Split groups
classical_models = ["Logistic Regression", "Tunned SVM", "Tunned KNN", "Random Forest"]
ensemble_models = ["XGBoost", "LightGBM", "Stacking Ensemble"]
df_classical = df.loc[classical_models].sort_values(by="Accuracy", ascending=False)
df_ensemble = df.loc[ensemble_models].sort_values(by="Accuracy", ascending=False)

# === Function to save individual plots ===
def save_table(df, title, filename):
    fig, ax = plt.subplots(figsize=(10, len(df) * 0.7 + 1))
    ax.axis('off')
    table = ax.table(cellText=df.values,
                     colLabels=df.columns,
                     rowLabels=df.index,
                     loc='center',
                     cellLoc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(12)
    table.scale(1.2, 2)
    ax.set_title(title, fontsize=16)
    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.close()

def save_heatmap(df, title, filename, cmap):
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.heatmap(df, annot=True, fmt=".2f", cmap=cmap, ax=ax, annot_kws={"size": 12})
    ax.set_title(title, fontsize=16)
    ax.tick_params(labelsize=12)
    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.close()

def save_bar(df, title, filename, palette=None):
    fig, ax = plt.subplots(figsize=(12, 6))
    df.plot(kind='bar', ax=ax, width=0.75, color=palette)
    ax.set_title(title, fontsize=16)
    ax.legend(loc='lower center', ncol=5, fontsize=12)
    ax.tick_params(axis='x', labelrotation=0, labelsize=12)
    ax.tick_params(axis='y', labelsize=12)
    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.close()

def save_combined_roc(model_files, filename):
    fig, ax = plt.subplots(figsize=(10, 8))
    for name, path in model_files.items():
        try:
            bundle = joblib.load(path)
            model = bundle["model"]
            if hasattr(model, "predict_proba"):
                y_prob = model.predict_proba(X_test)[:, 1]
                auc = roc_auc_score(y_test, y_prob)
                fpr, tpr, _ = roc_curve(y_test, y_prob)
                ax.plot(fpr, tpr, label=f"{name} (AUC = {auc * 100:.2f}%)")
        except Exception as e:
            print(f"ROC error for {name}: {e}")
    ax.plot([0, 1], [0, 1], 'k--', label="Chance")
    ax.set_title(" Combined ROC Curves for All Models", fontsize=16)
    ax.set_xlabel("False Positive Rate")
    ax.set_ylabel("True Positive Rate")
    ax.legend(loc="lower right", fontsize=10)
    ax.grid(True)
    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.close()

# === Save individual images ===
save_table(df_sorted, "All Models - Metrics Table", "all_models_table.png")
save_heatmap(df_sorted, " All Models - Heatmap", "all_models_heatmap.png", cmap="YlGnBu")
save_bar(df_sorted, " All Models - Bar Chart", "all_models_bar.png")

save_table(df_classical, " Classical Models - Table", "classical_models_table.png")
save_heatmap(df_classical, " Classical Models - Heatmap", "classical_models_heatmap.png", cmap="Blues")
save_bar(df_classical, "Classical Models - Bar Chart", "classical_models_bar.png", palette=sns.color_palette("Blues", n_colors=len(df_classical.columns)))

save_table(df_ensemble, " Ensemble Models - Table", "ensemble_models_table.png")
save_heatmap(df_ensemble, " Ensemble Models - Heatmap", "ensemble_models_heatmap.png", cmap="Oranges")
save_bar(df_ensemble, " Ensemble Models - Bar Chart", "ensemble_models_bar.png", palette=sns.color_palette("Oranges", n_colors=len(df_ensemble.columns)))

save_combined_roc(model_files, "all_models_roc_curve.png")

print(" All individual diagrams saved.")

# === Also save one big figure ===
fig, axes = plt.subplots(10, 1, figsize=(18, 70))

# Add all visuals into a single image
def embed_table(ax, df, title):
    ax.axis('off')
    tbl = ax.table(cellText=df.values,
                   colLabels=df.columns,
                   rowLabels=df.index,
                   loc='center',
                   cellLoc='center')
    tbl.auto_set_font_size(False)
    tbl.set_fontsize(12)
    tbl.scale(1.2, 2)
    ax.set_title(title, fontsize=18)

embed_table(axes[0], df_sorted, " All Models - Metrics Table")
sns.heatmap(df_sorted, annot=True, fmt=".2f", cmap="YlGnBu", ax=axes[1], annot_kws={"size": 12})
axes[1].set_title(" All Models - Heatmap", fontsize=16)

df_sorted.plot(kind='bar', ax=axes[2])
axes[2].set_title(" All Models - Bar Chart")

embed_table(axes[3], df_classical, "Classical Models - Table")
sns.heatmap(df_classical, annot=True, fmt=".2f", cmap="Blues", ax=axes[4])
axes[4].set_title(" Classical Models - Heatmap")

df_classical.plot(kind='bar', ax=axes[5], color=sns.color_palette("Blues", n_colors=len(df_classical.columns)))
axes[5].set_title("Classical Models - Bar Chart")

embed_table(axes[6], df_ensemble, " Ensemble Models - Table")
sns.heatmap(df_ensemble, annot=True, fmt=".2f", cmap="Oranges", ax=axes[7])
axes[7].set_title(" Ensemble Models - Heatmap")

df_ensemble.plot(kind='bar', ax=axes[8], color=sns.color_palette("Oranges", n_colors=len(df_ensemble.columns)))
axes[8].set_title(" Ensemble Models - Bar Chart")

# Combined ROC at the bottom
axes[9].set_title(" Combined ROC Curves for All Models", fontsize=16)
axes[9].set_xlabel("False Positive Rate", fontsize=12)
axes[9].set_ylabel("True Positive Rate", fontsize=12)

for name, path in model_files.items():
    try:
        bundle = joblib.load(path)
        model = bundle["model"]
        if hasattr(model, "predict_proba"):
            y_prob = model.predict_proba(X_test)[:, 1]
            auc = roc_auc_score(y_test, y_prob)
            fpr, tpr, _ = roc_curve(y_test, y_prob)
            axes[9].plot(fpr, tpr, label=f"{name} (AUC = {auc * 100:.2f}%)")
    except:
        continue

axes[9].plot([0, 1], [0, 1], 'k--', label="Chance")
axes[9].legend(loc="lower right", fontsize=10)
axes[9].grid(True)

plt.tight_layout()
plt.savefig("full_model_evaluation_with_roc.png", dpi=300, bbox_inches='tight')
plt.show()
print(" Saved full diagram: full_model_evaluation_with_roc.png")
```
